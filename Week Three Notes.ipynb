```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,'r--')
plt.xlabel('x')
plt.ylabel('y')
```

```python
plt.plot(x,y,color='purple', lw=2, ls='-',marker='o',markersize=20,markerfacecolor='yellow',
         markeredgewidth=3,markeredgecolor='green')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

for current_ax in axes:
    current_ax.plot(x,y,'r')
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2,figsize=(12,4))

axes[0].plot(x,y,'b')
axes[1].plot(y,x,'b--')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Big Plot')
axes1.set_xlabel('x')
axes1.set_ylabel('y')

axes2.plot(y,x)
axes2.set_title('Smaller Plot')
axes2.set_xlabel('y')
axes2.set_ylabel('x')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='x squared')
ax.plot(x,x**3,label='x cubed')

ax.legend(loc=0)
```

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'], bins=30)
```

```python
sns.distplot(tips['total_bill'], kde=False, bins=30)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
sns.pairplot(tips, hue='sex')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
sns.countplot(x='sex', data=tips)
```

```python
sns.countplot(x='smoker', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker',split=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex',dodge=True)
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
sns.factorplot(x='day', y='total_bill', data=tips, kind='bar')
```

```python
tc = tips.corr()
```

```python
sns.heatmap(tc, annot=True, cmap='coolwarm')
```

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
fp = flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
sns.heatmap(fp,cmap='coolwarm',linecolor='white',linewidths=0.1)
```

```python
sns.clustermap(fp,cmap='coolwarm', standard_scale=1)
```
